## Age

```python
mean_age = pd.concat([df_train['Age'], df_test['Age']]).mean()

nan_age_train_idx = pd.isnull(df_train['Age'])
df_train['Age_nan'] = nan_age_train_idx.astype(int)
df_train['Age'][nan_age_train_idx] = mean_age

nan_age_test_idx = pd.isnull(df_test['Age'])
df_test['Age_nan'] = nan_age_test_idx.astype(int)
df_test['Age'][nan_age_test_idx] = mean_age
```

## Embarked

```python
df_train = pd.get_dummies(df_train, columns=['Embarked'], dummy_na=True, drop_first=True)
df_test = pd.get_dummies(df_test, columns=['Embarked'], dummy_na=True, drop_first=True)
```

## Name

Extract title from name

```python
import re

def extract_title(name):
    m = re.search('[^,]+, ([^\.]+)\..*', name)
    return m.group(1)

df_train['Title'] = df_train['Name'].apply(extract_title) 
df_test['Title'] = df_test['Name'].apply(extract_title)

df_train = pd.get_dummies(df_train, columns=['Title'], dummy_na=True, drop_first=True)
df_test = pd.get_dummies(df_test, columns=['Title'], dummy_na=True, drop_first=True)
```
